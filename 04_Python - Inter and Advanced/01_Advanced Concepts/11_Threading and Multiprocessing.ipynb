## GIL (Global Interpreter Lock)
Python(CPython) has a mutex lock for every process called GIL that allows only one thread to access at one time. This is implemented as the memory management in CPython is not thread safe.

This means that threading in Python is not perfect like that of C++ and jave. 

Note: This doesnot mean that threading doesn't exist in python. The GIL only prevents multiple threads from accessing the python bytecode at once. So in other situations (like accessing I/O devices), Python threading works just fine. For CPU-bound tasks with a need for true parallelism, alternatives like the multiprocessing module or other concurrency models might be more suitable.

## Multiprocesssing

```python
from multiprocessing import Process
import os
import time

def p():
    for i in range(50):
        num = i**1000000
        time.sleep(.1)
        
process_list = []

num_cores = os.cpu_count()
print(num_cores)
# num_cores = 30
for i in range(num_cores):
    pro = Process(target=p)
    process_list.append(pro)
    
for p in process_list:
    p.start()   

for p in process_list:
    p.join()
    print("Process completed")
    
print("Completed")
```

## Threading

```python
from threading import Thread 
import os
import time

def p():
    for i in range(50):
        num = i**100000
        time.sleep(.1)
        
threads_list = []

num_threads = 100 # They are pretty lightweight
for i in range(num_threads):
    th = Thread(target=p)
    threads_list.append(th)
    
for t in threads_list:
    t.start()   

for t in threads_list:
    t.join()
print("Process completed")
```
